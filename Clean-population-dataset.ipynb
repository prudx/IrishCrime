```python
import pandas as pd

filename = 'PopulationData2002-2016-dirty.csv'

dataframe = pd.read_csv(filename)

dataframe
```

```python
df1 = pd.melt(dataframe, id_vars="County", var_name="Year", value_name="Population" )
df2 = df1.sort_values(by=['County','Year'], ascending=True)
df2
```
